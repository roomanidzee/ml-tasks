```python
import sys
sys.path.append('../..')
from src.commons.methods import generate_random_points
from src.mst.utils import create_point_structure
from src.mst.main import connect_structures, remove_longest_link
```

```python
import matplotlib.pyplot as plt
```

```python
test_data = generate_random_points(0.0, 100.0, 50)
points = create_point_structure(test_data)
```

```python
for p in points:
    plt.scatter(p.point_data.x, p.point_data.y, color='g')
```

```python
connected_links = connect_structures(points, len(points))
final_links = remove_longest_link(points, connected_links)
```

```python
for li in final_links:
    for p in points:
        if p.point_id == li.target_id:
            x2 = p.point_data.x
            y2 = p.point_data.y
    for p in points:
        if p.point_id == li.source_id:
            x1 = p.point_data.x
            y1 = p.point_data.y
    plt.plot([x1, x2], [y1, y2], color='g')
plt.show()    
```
